## invarients 
for the definitions of each invarients please refere to the obsidian vault on this project.
1) q-extent: with q=3.
2) q-extent: with q=5.
3) q-extent: with q=7.
4) excess-global
5) q-packing: with q=3.
6) q-packing: with q=5.
7) q-packing: with q=7.

```python
import scipy.sparse as sparse
import scipy.io as sio
import scipy.stats as stats
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# own code
import graph_tools as gtl
import invarients as inva

result_indecies = [
    "3-extend",
    "5-extend",
    "7-extend",
    "excess-global",
    "3-packing",
    "5-packing",
    "7-packing"
    ]
```

## power
### power-eris1176


```python
power_sci_array = sio.mmread("networks1/power/power-eris1176.mtx")
power_graph = nx.from_scipy_sparse_array(power_sci_array)
options = {
    "font_size": 20,
    "node_size": 5,
    "node_color": (0,0,0),
    "edge_color": "grey",
    "linewidths": 1,
    "width": 1,
    "with_labels" : False,

}

nx.draw_networkx(power_graph,**options)

```

```python
res_power_graph = pd.Series(None,result_indecies)
res_power_graph["3-extend"] = inva.q_extent_single(3,power_graph)
res_power_graph["5-extend"] = inva.q_extent_single(5,power_graph)
res_power_graph["7-extend"] = inva.q_extent_single(7,power_graph)
res_power_graph["excess-global"]  = inva.excess_global(power_graph)
res_power_graph["3-packing"] = inva.q_packing(power_graph,3)
res_power_graph["3-packing"] = inva.q_packing(power_graph,5)
res_power_graph["3-packing"] = inva.q_packing(power_graph,7)
```

```python
df = pd.DataFrame(res_power_graph)
df.to_csv("results/power-eris.csv")
```
